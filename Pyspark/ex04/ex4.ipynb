```python
import numpy as np
from numpy import linalg as L
```

```python
#page ranking function
def pageRanking(M, betha = 0.8, upsilon = .00001):
    A = np.zeros(np.shape(M))
    A = betha * M 
    ranking = np.ones(len(A))/ len(A)
    #temp = ranking.copy()
    count = 0
    threshold = upsilon + 0.1
    while threshold > upsilon:
        temp = ranking.copy()
        ranking = A.dot(ranking) + (1-betha)* (np.ones(len(A))/ len(A))
        #threshold = 0.0
        #temp = temp - ranking
        threshold = L.norm((temp - ranking), ord=1)
        count +=1
        
    return ranking,count # returning the ranking of each page and the number of iteration

# clique matrix fuction 
def clique(n=3):
    M = np.ones((n,n))/(n-1)
    np.fill_diagonal(M,0)
    return M
```

```python
#testing with a manually made example
M = np.array([[0,  1/2, 0 , 0],
              [1/3, 0 , 0 ,1/2],
              [1/3, 0 , 1 ,1/2],
              [1/3,1/2, 0 ,0]])
pageRanking(M, 0.8, 1/12)
```

```python
# creation of a clique with 4 vertices
M = clique(4)
pageRanking(M, 0.8, 1/12)
```

```python
# creation of a clique with 6 vertices
M = clique(6)
pageRanking(M, 0.8, 1/12)
```
